# 处理prompt文件夹下的所有markdown文件

````python
import re
import os

# 读取prompt文件夹下的所有文件
prompt_files = os.listdir("prompts")
print("prompt文件夹下的所有文件：")
print(prompt_files)
````

````python
for file in prompt_files:
    # 读取markdown文件中的内容
    with open(f"prompts/{file}", "r", encoding="utf-8") as f:
        markdown_content = f.read()

    pattern = r"```markdown\s*(.*?)\s*```"
    match = re.search(pattern, markdown_content, re.DOTALL)

    if match:
        extracted_content = match.group(1)
        print("提取的内容：")
        print(extracted_content)
        # 将提取的内容写入新文件
        with open(f"prompts/{re.sub(r"\.md$", "", file)}.txt", "w", encoding="utf-8") as f:
            f.write(extracted_content)
    else:
        print("没有找到匹配的内容。")
````

````python
# 删除prompt文件夹下的无后缀文件
for file in prompt_files:
    if "." not in file:
        os.remove(f"prompts/{file}")
        print(f"删除无后缀文件：{file}")
````
